```python
start_df_col = 0
end_df_col = 25
x_col = 0
col_num = 3
df_to_plot = x_pca


def plot_batch(df_to_plot, x_col, start_df_col, end_df_col, col_num):
    row_num = (end_df_col-start_df_col)//col_num + \
        ((end_df_col-start_df_col)/col_num > 0)
    fig = make_subplots(rows=row_num, cols=col_num)
    marker = go.scatter.Marker(size=2)
    df_col = start_df_col
    for row_pointer in range(1, row_num+1):
        for col_pointer in range(1, col_num+1):
            if df_col > end_df_col:
                continue
            fig.add_trace(
                go.Scatter(
                    x=df_to_plot.iloc[:, x_col],
                    y=df_to_plot.iloc[:, df_col],
                    mode='markers',
                    marker=marker),
                row=row_pointer, col=col_pointer,
            )
            df_col += 1
    fig.update_layout(height=1500, width=1000,
                      title_text="X vs Y", showlegend=False)
    fig.show()
```
