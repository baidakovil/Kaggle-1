```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import metrics
from sklearn.cluster import AffinityPropagation
```

```python
# читаем данные и заносим в pd.DataFrame без индексной колонки
df = pd.read_csv('incoming/data.csv')
df.drop('id', axis=1, inplace=True)
X = df.copy()
```

```python
# переводим весь массив в int8
# здесь только округляем
Xint8 = X.select_dtypes(include='int64').astype('int8', copy=False)
X.loc[:,'f_07':'f_13']=Xint8
# здесь умножаем на 10 и округляем
Xint9 = X.select_dtypes(include='float64')*10
Xint9 = Xint9.astype('int8', copy=False)
X.loc[:,'f_00':'f_06']=Xint9.loc[:,'f_00':'f_06']
X.loc[:,'f_14':'f_28']=Xint9.loc[:,'f_14':'f_28']
```
